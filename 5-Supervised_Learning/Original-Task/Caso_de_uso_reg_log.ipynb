```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

path = '/content/drive/MyDrive/Diplomado data science/V. Aprendizaje supervisado/files'

df = pd.read_csv(path+'/titanic.csv', low_memory=False)
df.head()
```

```python
half_count = len(df) / 2
df = df.dropna(thresh=half_count,axis=1) # Drop any column with more than 50% missing values
df = df.drop(['PassengerId','Name','SibSp','Parch','Ticket', 'Signing_date' ],axis=1)
```

```python
df = df.drop(['Fare'],axis=1)
```

```python
df = df.drop(['Embarked'],axis=1)
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

le = LabelEncoder()
df["Sex"] = le.fit_transform(df.Sex)
```

```python
df.head()
```

```python
caract = ["Pclass", "Sex", "Age"]

y = df.pop("Survived")
X = df[caract]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```
