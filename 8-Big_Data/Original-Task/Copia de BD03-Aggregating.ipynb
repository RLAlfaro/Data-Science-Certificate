```python
import pyspark.sql.functions as f
```

```python
from pyspark.sql.functions import countDistinct, avg, stddev
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
path = "/content/drive/MyDrive/Cursos/Data Science UDD/Big Data/sales_info.csv"
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").getOrCreate()
```

```python
df = spark.read.csv(path, header=True)
df.show()
```

```python
df.groupBy("Company").count().show()
```

```python
sales = df.groupBy("Company").agg(f.max("Sales").alias("Max"))
sales.show()
```

```python
sales.orderBy("Max", ascending=False).show()
```

```python
df1 = df.describe().select('Summary', 'Sales')
df1.filter(df1.Summary.isin(['mean', 'stddev'])).show()
```

```python
df2 = sales.describe().select('Summary', 'Max')
df2.filter(df2.Summary.isin(['mean', 'stddev'])).show()
```
