```
patientInfoDF.createOrReplaceTempView("patients")
```

```
%sql select sex,city, count(*) from patients where sex is not null group by sex,city
```

```
patientInfoDF.select('infection_case').distinct().count()
```

###Infection reasons for the patients

```
%sql select infection_case,city,count(*)  from patients where infection_case is not null group by infection_case,city
```

```
%sql select confirmed_date, count(patient_id) from patients where confirmed_date is not null group by confirmed_date order by confirmed_date;
```

###Recovered Patients in each city

```
%sql
select city, count(patient_id) from patients  where released_date is not null group by city order by count(patient_id) desc
```

```
%sql select sex,city, count(*) from patients where sex is not null and city == "Gyeongsan-si" group by sex,city
```

###Fatalities by City

```
%sql
select city, count(patient_id) from patients  where deceased_date is not null and city is not null group by city order by count(patient_id) desc
```

###% of Recovered Patient cases by type of Infection

```
%sql

select infection_case,count(patient_id) from patients where released_date is not null and infection_case is not null group by infection_case 
```

###% of Fatality cases by type Infection
This graph shows the recovery percentage of deaths from different types of infections. It mainly shows the fatalities with "unknown reason". We can assmue a lot of factors for fatalities but we may need more data or insights.

```
%sql
select infection_case, count(patient_id) from patients where deceased_date is not null and infection_case is not null group by infection_case
```
