## <font color='blue'>**Medidas de Tendencia Central**</font>

Las medidas de tendencia central son  valores que se encuentra hacia el centro de la distribución de los valores de una serie de observaciones.

La **Moda** de una variable Cualitativa o Cuantitativa, es el valor que se presenta con mayor frecuencia en una distribución de datos.

1. Cuando dos datos se presentan con la misma frecuencia y esta es la más alta, ambos datos son moda, es decir,  la distribución es bimodal.
2. Cuando dos o más datos se presentan con la misma frecuencia y esta es la más alta, todos datos son moda, es decir, la distribución es multimodal.
3. Cuando todos los datos tienen la misma frecuencia, se dice que no hay moda.

La **Mediana** de una variable cualitativa ordinal o cuantitativa, corresponde al valor intermedio, cuando los valores de la variable se presentan en orden de magnitud creciente (o decreciente).


$$ Me = X_{\frac{n}{2}}, \mbox{  Para el caso que n es impar.}  $$ 

$$ Me = \frac{X_{\frac{n}{2}} + X_{\frac{n}{2}+1}}{2},  \mbox{  Para el caso que n es par.}$$ 

La **media aritmética** de una variable cuantitativa, se calcula sumando los valores y dividiéndolos por el total de ellos.

$$ Media = \frac{\sum X_i}{n}$$





## <font color='blue'>**Medidas de dispersión**</font>

![tipos de variables](https://drive.google.com/uc?export=view&id=1cWkiQdRPr239DmRtsgTxkQ2fqtBcoWPQ)


**Rango**: No es una medida significativa, pues solo indica cuán dispersos están los datos entre los extremos. 

$$R = X_{max} - X_{min}$$

**Coeficiente de variación**: También llamado _Coeficiente de Variación de Pearson_, es una medida estadística que nos informa acerca de la dispersión relativa de un conjunto de datos. Es decir, nos informa al igual que otras medidas de dispersión, de si una variable se mueve mucho, poco, más o menos que otra. Se utiliza cuando se necesita comparar la dispersión de distribuciones que tienen distintas unidades. 

$$CV = \frac{S}{\bar{X}}$$

Por ejempo, se quiere comprar la dispersión entre la altura de 50 alumnos de una clase y su peso. Para comparar la altura podríamos utilizar como unidad de medida metros y centímetros y para el peso el kilogramo. Comparar estas dos distribuciones mediante la desviación estándar, no tendría sentido dado que se pretenden medir dos variables cuantitativas distintas (una medida de longitud y una de masa).

## <font color='blue'>**Medidas de posición**</font>

**Cuartiles**  son medidas de posición  que dividen  la distribución ordenada en cuatro partes iguales, ellos son $Q_1 , Q_2 , Q_3 $.

$Q_1$: ***Cuartil 1***. Es el valor de la variable que deja a la izquierda el 25% de los datos.

**Rango intercuartilico (IQR)**: es una estimación estadística de dispersión  de una distribución de datos y corresponde a:  

$$IQR = Q_3-Q_1$$

**Percentiles** son medidas de posición que dividen la distribución ordenada en 100 partes iguales, ellos son $P_1$, $P_2$, ..., $P_k$ , con $k=[1,99]$.

$P_K$ :  ***Percentil k***. Es el valor de la variable que deja a la izquierda el $k\%$ de los datos.

### Parámetros (poblacionales)
Dijimos que un *parámetro* es un valor que caractariza un aspecto de la población; e.g., media poblacional, mediana poblacional, desviación estándar poblacional, etc.

Analicemos los parámetros de la población **poblacion**. Como los datos no están guardados en una estructura Pandas (Serie, DataFrame), utilizaremos las librerías _Nympy_ y _scipy_ para calcular parámetros y estadísticos.

`````python
from scipy import stats # importando scipy.stats
#Documentación: https://docs.scipy.org/doc/scipy/reference/stats.html

print(f'Tamaño población\t\t= {len(poblacion):d} individuos')
print(f'Media poblacional\t\t= {poblacion.mean():4.2f}')
print(f'Desviación estándar poblacional\t= {poblacion.std():4.2f}')
print(f'Mediana poblacional\t\t= {np.median(poblacion):4.2f}')
`````

Ahora analicemos los parámetros de la población **edades**. Incorporamos una nueva medida, la __moda__. La moda es el valor que aparece con mayor frecuencia dentro de un conglomerado. En un grupo puede haber una única moda (`unimodal`), dos modas (`bimodal`), o más de dos modas (`multimodal`). Se llama a un grupo `amodal` cuando en él no se repiten los valores.

Por último, se conoce como `moda adyacente` cuando dos valores continuos tienen la misma cantidad de repeticiones. En este caso se saca el promedio de ambos.

`````python
print(f'Tamaño población edades\t\t\t= {len(edades):d} individuos')
print(f'Media de edades\t\t\t\t= {edades.mean():4.2f}')
print(f'Desviación estándar de las edades\t= {edades.std():4.2f}')
print(f'Mediana poblacional de las edades\t= {np.median(edades):4.2f}')
print(f'Moda poblacional es {stats.mode(edades)[0][0]} y se repite {stats.mode(edades)[1][0]} veces')
`````

### Estadísticos (muestrales)
Medida descriptiva de la muestra que se utiliza para estimar al respectivo parámetro poblacional. El *estadístico* es básicamente lo mismo que el *parámetro poblacional*, pero se diferencia en que se calcula a partir de la muestra y no de la población. <br>
La desvisción estándar para muestras se denomina "__desviación estándar insesgada__" ya que se calcula dividiendo por $(n - 1)$.

**Observación**

La desviación estándar muestral se denota con el símbolo: $S$ o $\widehat{\sigma}$

La media muestral se denota con el símbolo: $\overline{X}$ o $\widehat{\mu}$

## <font color='blue'>**Box Plots**</font>

Un "Box plot" es una forma de representar estadísticamente la *distribución* de los datos a través de cinco dimensiones principales:

- **Minimun:** Número más pequeño en el conjunto de datos.
- **First quartile:** Número medio entre el `mínimo` y la` mediana`.
- **Second quartile (Median):** Número medio del conjunto de datos (ordenado).
- **Third quartile:** Número medio entre `mediana` y` máximo`.
- **Maximum:** Número más alto en el conjunto de datos.
