## <font color='blue'>**Definiciones**</font>

Una hip√≥tesis estad√≠stica es una afirmaci√≥n o proposici√≥n respecto a alguna caracter√≠stica de una poblaci√≥n, generalmente fundamentada sobre un par√°metro de la misma. Contrastar una hip√≥tesis es comparar las predicciones con la realidad que observamos ocurrida en una muestra. Si dentro del margen de error que estamos dispuestos a admitir, hay coincidencia, aceptaremos la hip√≥tesis y en caso contrario la rechazaremos.

1. La hip√≥tesis emitida se suele designar por¬†$ùëØ_ùüé$¬†y se llama¬†**Hip√≥tesis nula**. Lo de ‚Äúnula‚Äù viene de que partimos del supuesto de que las diferencias entre el valor verdadero del par√°metro y su valor hipot√©tico, en realidad no son tales sino debidas al azar, es decir, no hay diferencia, o dicho de otra forma la diferencia es nula.

2. La hip√≥tesis contraria se designa por¬†$ùëØ_ùüè$¬†¬†y se llama¬†**Hip√≥tesis alternativa**.

### Hipotesis nula

La hip√≥tesis nula (denotada por $H_0$) es la afirmaci√≥n de que el valor de un par√°metro de la poblaci√≥n (media, proporci√≥n o desviaci√≥n est√°ndar) es igual a un valor aseverado.

Las siguientes son hip√≥tesis nulas t√≠picas:

1. $H_0: \mu = 37$
2. $H_0: p = 0.5$
2. $H_0: \sigma = 15$

La hip√≥tesis nulas se aprueba en forma directa, en el sentido de que **se supone que es verdadera**, y llegamos a una conclusi√≥n de rechazar $H_0$ o no rechazar $H_0$.

### Hipotesis alternativa 

La hip√≥tesis alternativa (denotada por $H_1$) es la afirmaci√≥n de que el par√°metro tiene un valor que difiere de la hip√≥tesis nula. Los s√≠mbolos que emplea la hip√≥tesis alternativa son: $\lt$, $\gt$ o bien $\ne$.

![Matriz Covarianza](https://drive.google.com/uc?export=view&id=1ZKsYo3RhWZ5jckX0Zc6YkSeeGW8O9tkz)

Si est√° realizando un estudio y desea emplear una prueba de hip√≥tesis para sustentar una aseveraci√≥n, esta debe redactarse de tal manera que se convierta en la hip√≥tesis alternativa. Esto quiere decir, que su aseveraci√≥n debe expresarse utilizando solo estos s√≠mbolos $\lt$, $\gt$ o bien $\ne$.




### Ejemplos

1. Sospechamos que las bolsas de frutos secos de 100 gramos, realmente no pesan 100 gramos. Para contrastar esta hip√≥tesis plantear√≠amos:

$$H_0: \mu = 100$$
$$H_1: \mu \neq 100$$

2. Pensamos que la proporci√≥n de gente que vot√≥ al partido A en las elecciones (35%) ahora es inferior ya que no lo han hecho muy bien. Para contrastar esta hip√≥tesis:

$$H_0:p \geq 0.35 $$
$$H_1:p < 0.35 $$

3. Estar√≠a contento de comprobar que no pueden demostrar que mi media de notas ha bajado de 6.2 como parecen indicar los √∫ltimos ex√°menes. Para contrastar esta hip√≥tesis:

$$H_0:\mu \leq 6.2 $$
$$H_1:\mu > 6.2 $$

### Contraste de hip√≥tesis
Se formula la hip√≥tesis sobre la poblaci√≥n (ej. $\mu$, $p$, $\sigma$, etc.).  Entonces las conclusiones sobre la validez de la hip√≥tesis se basar√°n en la informaci√≥n de una muestra (ej. $\bar{x}$, $\hat{p}$, $S$, etc).

La realizaci√≥n de un contraste de hip√≥tesis __no consiste en__ decidir cu√°l de las dos hip√≥tesis ($H_0$ o $H_1$) es m√°s creible, sino en decidir si la muestra proporciona o no suficiente evidencia para descartar $H_0$. Mientras no tengamos suficiente evidencia para aceptar $H_1$ tendremos que creernos que $H_0$ es cierta.

## <font color='blue'>**Estad√≠stico de prueba**</font>
El **estad√≠stico de prueba**, es un valor que se utiliza para tomar la decisi√≥n sobre la hip√≥tesis nula y se calcula convirtiendo el **estad√≠stico muestral** (tales como la media muestral $\bar{x}$, o la proporci√≥n muestral $\hat{p}$) en una puntuaci√≥n (tales como $Z$ o $t$), bajo el supuesto que la hip√≥tesis nula es verdadera.

### Prueba $Z$

La __prueba $Z$__, es un tipo de prueba param√©trica que se utiliza para determinar el grado de significatividad estad√≠stica de las diferencias entre las medias de dos conjuntos de datos. Se utiliza cuando las muestras son amplias (de m√°s de 30 individuos aproximadamente) e independientes (esto es, no correlacionadas). Opera con la media, la desviaci√≥n t√≠pica y el n√∫mero de individuos de cada muestra. Una vez calculado el valor de $Z$, se comparar√° con el valor cr√≠tico de $Z$ obtenido de una curva de distribuci√≥n normal. La hip√≥tesis nula sostiene que la diferencia entre medias no es significativa y ser√° rechazada si el valor obtenido en la prueba Z sobrepasa el valor cr√≠tico.

El estad√≠stico de __prueba $Z$__, es un valor que se utiliza para tomar la decisi√≥n sobre la hip√≥tesis nula y se calcula convirtiendo el estad√≠stico muestral (como la media muestral $\bar{x}$, proporci√≥n muestral $\hat{p}$) en una puntuaci√≥n (como $Z$), bajo el supuesto que la hip√≥tesis nula es verdadera. Este test considera que $\sigma$,  (Desviacion est√°ndar) es conocida.

### Valor-P o P-Value
El __nivel cr√≠tio $P$__(o __Valor-$P$__), es el nivel de significaci√≥n (1 - nivel de confianza) __m√°s peque√±o__ para el que la muestra obtenida obligar√≠a a rechazar la hip√≥tesis nula.
* El Valor-P es la probabilidad (calculada bajo $H_0$) de obtener un resultado que sea menos compatible con $H_0$ que el resultado obtenido con la muestra.
* El Valor-P depende de la muestra.
* Puede considerarse como el apoyo que la hip√≥tesis nula recibe de las observaciones de la muestra:
 * Si el Valor-P e __menor__ que el nivel de significaci√≥n ($\alpha$) prefijado, el apoyo a $H_0$ es escaso y, por tanto, $H_0$ debe rechazarse.
 * Si el Valor-P es __mayor o igual__ que el nivel de significaci√≥n ($\alpha$) prefijado, el apoyo a $H_0$ es suficiente y, por tanto, $H_0$ __NO__ debe rechazarse.

### Criterios de decisi√≥n
El procedimiento convencional de prueba de hip√≥tesis requiere que __siempre probemos la hip√≥tesis nula__, de manera que nuestra conclusi√≥n inicial siempre ser√° una de las siguientes:

1. Rechazo la Hip√≥tesis nula.
2. No rechazo la hip√≥tesis nula.

La decisi√≥n de rechazar o no una hip√≥tesis nula, se puede realizar a trav√©s de dos m√©todos; el __M√©todo cl√°sico__ o por __Valor P__:

__M√©todo cl√°sico__<br>
* __Rechazo__ $H_0$  si el estad√≠stico de prueba ($Z$ o $t$) cae dentro de la Regi√≥n cr√≠tica.

* __No Rechazo__ $H_0$  si el estad√≠stico de prueba __NO__ cae dentro de la Regi√≥n cr√≠tica.

__Valor P__<br>
* __Rechazo__ $H_0$  si el valor $P \le \alpha$, donde $\alpha$ es el nivel de significancia.
* __No Rechazo__ $H_0$  si el valor $P \gt \alpha$

```
# Tu c√≥digo aqu√≠ ...

# X = la talla media de los hombres de 18 a√±os o m√°s de un pais es igual a 180 (con un 5% de significancia)
# X‚àº(Œº,4) 

import pandas as pd

alt_muestra = [167, 167, 168, 168, 168, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195]
x = pd.Series(alt_muestra)

# 1. Calcular estadisticos muestra

Mu = x.describe()['mean']
Sig = 4
S   =  x.describe()['std']
n = x.describe()['count']
print(Mu, Sig, S, n)
             
```

### **T-Test**

__Test-T__ es una prueba en la que el estad√≠stico utilizado tiene una distribuci√≥n __t de Student__ si la hip√≥tesis nula es cierta. Se aplica cuando la poblaci√≥n estudiada sigue una distribuci√≥n normal, pero el tama√±o muestral es demasiado peque√±o como para que el estad√≠stico en el que est√° basada la inferencia est√© normalmente distribuido, utiliz√°ndose una estimaci√≥n de la desviaci√≥n t√≠pica en lugar del valor real.

Para calcular __t__ se utiliza la siguiente formula: 

$$ t = \frac{\bar{x} - \mu}{\frac{S}{\sqrt{n}}}$$

Donde $S$ es la la desviacion estandar muestral _insesgada_. 

### Ejemplo: Hip√≥tesis para la media  ($\sigma$ desconocido)
### Rentabilidad fondos de inversi√≥n

Sea X la variable ‚Äùrentabilidad de cierto tipo de fondos de inversi√≥n tras una apreciaci√≥n fuerte del marco con respecto al d√≥lar‚Äù. Se considera que la media de esta variable es 15. Un economista afirma que dicha rentabilidad media ha variado, por lo que lleva a cabo un estudio en las condiciones rese√±adas anteriormente, sobre una muestra de 9 fondos cuya media muestral resulta ser de 15,308 y cuya varianza muestral es 0,193.

1. Especificando las hip√≥tesis necesarias, contrastar la afirmaci√≥n del economista al 5%.

2. A partir del resultado de punto 1, razonar si el intervalo de confianza para la media al 95% contendr√° o no al valor 15.

```
import numpy as np # importando numpy
from scipy import stats # importando scipy.stats
import matplotlib.pyplot as plt #importando matplotlib

# 1.  calcular estadisticos muestra
xbar = 15.308
S2 = 0.193 # varianza muestral
S = np.sqrt(S2) # desv. std muestral insesgada (asumimos)
n = 9

print(f'Xbarra: {xbar:4.2f}')
print(f'S: {S:4.2f}')
print(f'Tama√±o muestra: {n}')


# 2. Definir hipotesis a contrastar (prueba bilateral - 2 colas)
#    H0:   Mu = 15
#    H1:   Mu ‚â† 15

# 3. como n < 30 y no se conoce Sigma, aplicaremos T Student
alpha = 0.05
mu = 15

# 4. calcular valor critico de T
#    n-1 grados de libertad, donde n corresponde al numero de la muestra y 2 colas
Tcritico = stats.t.ppf(1-(alpha/2), df=(n-1)) 
print(f'T critico: {Tcritico:4.2f}')


# 5. calcular Estadistico de prueba t
t =  (xbar - mu) / (S / np.sqrt(n))
print(f't: {t:4.2f}')

# 6. Define Regla de decision
#    no rechaza H0 si: -Tcritico <= t <= Tcritico
#    rechazar H0 si:  t < -Tcritico  o  t > Tcritico  


# 7. calcular valor-p
pvalue = 1 - (stats.t.cdf(t, df=(n - 1)))
print(f'Valor-p: {pvalue*2:4.2f}') #2 colas
```

Conclusi√≥n:  

Como $t$ de prueba se encuentra entre $+- Tcritico$ , no se rechaza la $H_0$. No existe evidencia para afirmar que la rentabilidad ha cambiado.

Al observar el valor-p , este es mayor a $\alpha=0,05$. Por tanto no se rechaza $H_0$.

Utilice los datos muestrales con un nivel de significancia de 0,05, para probar la aseveraci√≥n que hizo un gerente de producci√≥n que los M&M tienen en realidad una media mayor que 0,8535 g. de manera que los consumidores est√°n recibiendo m√°s que la cantidad indicada en la etiqueta. 

dulces = (0.751, 0.841, 0.856, 0.799, 0.966, 0.859, 0.857, 0.942, 0.873, 0.809, 0.890, 0.878, 0.905)

1. Calcular estad√≠sticos de muestra

```
# Tu c√≥digo aqu√≠ ...

import numpy as np # importando numpy
from scipy import stats # importando scipy.stats
import matplotlib.pyplot as plt #importando matplotlib
import pandas as pd

dulces_list = [0.751, 0.841, 0.856, 0.799, 0.966, 0.859, 0.857, 0.942, 0.873, 0.809, 0.890, 0.878, 0.905]
x = pd.Series(dulces_list)
# print(x.describe()['mean'])

# 1.  calcular estadisticos muestra
xbar = x.describe()['mean']
S = x.describe()['std'] #np.sqrt(S2) # desv. std muestral insesgada (asumimos)
S2 = S**2 # varianza muestral
n = x.describe()['count']

print(10*'----')
print('\tEstadisticos de la muestra')
print(f'Xbarra: {xbar:4.4f}')
print(f'S: {S:4.4f}')
print(f'Tama√±o muestra: {n}')
print(10*'----')

# 2. Definir hipotesis a contrastar (prueba UNILATERIAL - 1 cola izquierda)
#    H0:   Mu >= 0.8535
#    H1:   Mu <  0.8535

# 3. como n < 30 y no se conoce Sigma, aplicaremos T Student
alpha = 0.05
mu = 0.8535 #esta es la hipotesis del gerente

## hasta ac√°

# 4. calcular valor critico de T
#    n-1 grados de libertad, donde n corresponde al numero de la muestra y 1 cola
Tcritico = stats.t.ppf(1-(alpha), df=(n-1)) 
print(f'T critico: {Tcritico:4.2f}')

# 5. calcular Estadistico de prueba t
t =  (xbar - mu) / (S / np.sqrt(n))
print(f't: {t:4.2f}')

# 6. Define Regla de decision
#    no rechazar H0 si: t > Tcritico
#    rechazar H0 si: t <= Tcritico 

# 7. calcular valor-p
pvalue = 1 - (stats.t.cdf(t, df=(n - 1)))
print(f'Valor-p: {pvalue:4.2f}') #1 cola
```

**_Conclusi√≥n:_** Se rechaza H_0 (por que t=0.63<1.78=T_critico), con un 95% de significancia no se puede afirmar que el contenido de una bolsa de M&M ser√° mayor que 396.9 gramos.

## <font color='blue'>**Errores Tipo I y II**</font>

Cuando probamos la  hip√≥tesis llegamos a la conclusi√≥n de rechazarla o no rechazarla. Tales conclusiones pueden ser correctas o incorrectas (incluso cuando se hace todo correctamente)ncia del contraste se acepta.

Ninguna prueba de hip√≥tesis es 100% cierta. Puesto que la prueba se basa en probabilidades, siempre existe la posibilidad de llegar a una conclusi√≥n incorrecta. Cuando se realiza una prueba de hip√≥tesis, se puede cometer dos tipos de error: __tipo I__ y __tipo II__. Los riesgos de estos dos errores est√°n inversamente relacionados y se determinan seg√∫n el nivel de significancia y la potencia de la prueba. Por lo tanto, se debe determinar qu√© error tiene consecuencias m√°s graves para su situaci√≥n antes de definir los riesgos.

### Error de tipo I
Si se rechaza la hip√≥tesis nula __$H_0$__ cuando es verdadera, se comete un __error de tipo I__. La probabilidad de cometer un __error de tipo I__ es __$\alpha$__, que es el __nivel de significancia__ que se establece para la prueba de hip√≥tesis. Un $\alpha = 0.05$ indica que se est√° dispuesto a aceptar una probabilidad de 5% de estar equivocado al rechazar la hip√≥tesis nula. Para reducir este riesgo, se debe utilizar un valor menor para $\alpha$. Sin embargo, usar un valor menor para alfa significa que usted tendr√° menos probabilidad de detectar una diferencia si esta realmente existe.

### Error de tipo II
Cuando la hip√≥tesis nula __$H_0$__ es falsa y no se la rechaza, se comete un __error de tipo II__. La probabilidad de cometer un error de __tipo II__ es __$\beta$__, la cual depende de la potencia de la prueba. Se reducir el riesgo de cometer un error de tipo II al asegurarse de que la prueba tenga suficiente potencia. Para ello, debemos asegurarnos de que el tama√±o de la muestra sea lo suficientemente grande como para detectar una diferencia pr√°ctica cuando esta realmente exista.

La probabilidad de rechazar la hip√≥tesis nula cuando es falsa es igual a $1 - \beta$. Este valor es la __potencia de la prueba__.

Verdad acerca de la poblaci√≥n

| Decisi√≥n basada <br> en la muestra | $H_0$ es verdadera | $H_0$ es falsa |
|-------------------------------|:---------------:|:-----------:|
| No rechazar $H_0$  | <font color='green'>__Decisi√≥n correcta__</font> <br>(probabilidad = $1-\alpha$)     |__Error tipo II__ <br> no rechazar $H_0$ cuando es falsa <br>(probabilidad = $\beta$) |
| Rechazar $H_0$   | __Error tipo I__ <br>rechazar $H_0$ cuando es verdadera <br>(probabilidad = $\alpha$)  | <font color='green'>__Decisi√≥n correcta__</font> <br>(probabilidad = $1-\beta$)|


### Ejemplo de error de tipo I y tipo II
Para entender la interrelaci√≥n entre los errores de __tipo I__ y __tipo II__, y para determinar cu√°l error tiene consecuencias m√°s graves para una situaci√≥n espec√≠fica, consideremos el siguiente ejemplo.

Un investigador m√©dico desea comparar la efectividad de dos medicamentos. Las hip√≥tesis nula y alternativa son:

Hip√≥tesis nula ($H_0$): $\mu_1 = \mu_2$, los dos medicamentos tienen la misma eficacia.

Hip√≥tesis alternativa ($H_1$): $\mu_1 \ne \mu_2$, los dos medicamentos no tienen la misma eficacia.

Un __error de tipo I__ se produce si el investigador rechaza la hip√≥tesis nula y concluye que los dos medicamentos son diferentes cuando, en realidad, no lo son. Si los medicamentos tienen la misma eficacia, el investigador podr√≠a considerar que este error no es muy grave, porque de todos modos los pacientes se beneficiar√≠an con el mismo nivel de eficacia independientemente del medicamento que tomen. Sin embargo, si se produce un __error de tipo II__, el investigador no rechaza la hip√≥tesis nula cuando debe rechazarla. Es decir, el investigador concluye que los medicamentos son iguales cuando en realidad son diferentes. Este error puede poner en riesgo la vida de los pacientes si se pone en venta el medicamento menos efectivo en lugar del medicamento m√°s efectivo.

Cuando realice las pruebas de hip√≥tesis, considere los riesgos de cometer errores de tipo I y tipo II. Si las consecuencias de cometer un tipo de error son m√°s graves o costosas que cometer el otro tipo de error, entonces elija un nivel de significancia y una potencia para la prueba que reflejen la gravedad relativa de esas consecuencias.
