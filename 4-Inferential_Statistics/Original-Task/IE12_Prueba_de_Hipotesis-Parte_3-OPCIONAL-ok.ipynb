## <font color='blue'>**A/B Test**</font>

Las pruebas A / B son una de las herramientas más importantes para optimizar la mayoría de las cosas con las que interactuamos en nuestras computadoras, teléfonos y tabletas. Desde diseños de sitios web hasta anuncios en redes sociales y características de productos, cada botón, banner y llamado a la acción probablemente haya sido probado A / B. Y estas pruebas pueden ser extremadamente granulares; Google probó la famosa prueba de “40 tonos de azul” para decidir qué tono de azul se debe usar para los enlaces en las páginas de destino de Google y Gmail.

### ¿Qué son las pruebas A / B?
Las pruebas A / B son una metodología general que se utiliza en línea cuando se prueban cambios de productos y nuevas funciones. Usted toma dos grupos de usuarios, muestra a un grupo de usuarios el producto modificado (grupo de experimentos) y muestra al segundo grupo de usuarios el producto original o el grupo de características (grupo de control). Luego, compara los dos grupos para determinar qué versión del producto es mejor.

Las pruebas A / B son importantes porque:

1. Elimina la necesidad de adivinar o confiar en la intuición.
2. Proporciona respuestas precisas rápidamente
3. Permite una iteración rápida de ideas
4. Establece relaciones causales, no solo correlaciones

Las pruebas A / B funcionan mejor cuando se prueban cambios incrementales, como cambios de User eXperience, nuevas funciones, clasificación y tiempos de carga de la página. Por lo general, es necesario que exista una cantidad importante de usuarios que permitan la inferencia estadística y los experimentos que se puedan completar en un período de tiempo razonable.

Las pruebas A / B no funcionan bien cuando se prueban cambios importantes, como nuevos productos, nuevas marcas o experiencias de usuario completamente nuevas. En estos casos, puede haber efectos novedosos que generen un compromiso más alto de lo normal o respuestas emocionales que hagan que los usuarios se resistan al cambio inicialmente. El cambio de marca de Instagram es un gran ejemplo de un mal caso de uso para las pruebas A / B que inicialmente causó una gran respuesta emocional negativa. Además, las pruebas A / B en productos que se compran con poca frecuencia y que tienen plazos de entrega prolongados son una mala idea: esperar a medir la tasa de recompra de las casas que compran los clientes va a arruinar los plazos de los experimentos.

### Investigación

Antes de sumergirse en cualquier experimento, investigue un poco. La investigación puede ayudarlo a diseñar un experimento, asegurarse de que utiliza las mejores prácticas e incluso puede descubrir resultados de experimentos similares. Hay varias fuentes de datos útiles que pueden ayudar a diseñar o respaldar un experimento:

Investigación y datos externos: el adagio "si lo has pensado, probablemente alguien lo haya probado" se aplica a los experimentos; Investigar un poco puede revelar los conocimientos que necesita o al menos proporcionar ideas para el diseño de experimentos.

Datos históricos: el análisis retrospectivo es barato y fácil y puede proporcionar información sobre las correlaciones históricas (¡pero no la causalidad!)

Investigación sobre la experiencia del usuario: poder ver realmente cómo un usuario interactúa con un producto puede ser invaluable para configurar y probar un experimento.

Grupos de enfoque: si bien no es accesible para la mayoría de los analistas, los grupos de enfoque pueden ser excelentes para probar cambios importantes y nuevos productos.

Encuestas: las encuestas pueden ser efectivas para recopilar información que no se revela en línea.

### Eleccion de las métricas

La elección de métricas es, como era de esperar, fundamental para el éxito de las pruebas A / B. Muchos experimentos comunes tienen métricas claras y de mejores prácticas para usar; sin embargo, cada empresa se enfrenta a sus propios matices y prioridades que deben tenerse en cuenta al elegir métricas; El desarrollo de métricas debe aprovechar la experiencia, el conocimiento del dominio y el análisis de datos exploratorios. En general las métricas funcionan mejor cuando se bajan en cascada de arriba hacia abajo, es decir, desde la estrategia de una empresa o unidad de negocio. De esta forma existe un vínculo claro entre los objetivos del experimento y los objetivos de la empresa. Este enfoque significa que los experimentos pueden basarse en una o dos métricas claras (por ejemplo, conversiones de clientes o número de usuarios activos) que son fáciles de informar y comprensibles para todas las partes interesadas.

Sin embargo, tener una o dos métricas clave no suele ser suficiente. La mayoría de los tomadores de decisiones querrán comprender qué ha impulsado el cambio, por lo que las métricas clave deben estar respaldadas por métricas más detalladas que ayuden a explicar los resultados. Un marco que puede proporcionar esta explicabilidad es el seguimiento de las métricas a lo largo del recorrido del usuario. Tener métricas en estos marcos ayuda a identificar claramente dónde se han producido los cambios y dónde se podría centrar la experimentación y la atención adicionales.

Se deben considerar cuatro categorías de métricas al diseñar un experimento, a saber:
1. Sumas y recuentos, p. Ej. cuántas cookies visitan una página web.

2. Métricas  como media, mediana y cuartiles, p. Ej. cuál es el tiempo medio de carga de la página.

3. Probabilidad y tasas, p. Ej. tasas de clics.

4. Razones, p. Ej. relación entre la probabilidad de que un clic genere ingresos y la probabilidad de cualquier clic.

### Detallar métricas

Una vez que haya decidido un conjunto de métricas, deberá detallarlas. Por ejemplo, la cantidad de usuarios activos es una métrica central utilizada por la mayoría de las empresas de tecnología, pero ¿qué significa esto realmente? Al definir esta métrica, algunas consideraciones incluirían:

¿Cuál es el período de tiempo para que un usuario esté activo? ¿Hay alguien activo que haya usado el producto hoy, en los últimos 7 días, en el último mes?
¿Qué actividad constituye activo? ¿Está iniciando sesión, pasando una cierta cantidad de tiempo usando el producto? ¿Qué pasa si la actividad es solo una notificación automática?
¿Importa qué tan activo es el usuario, p. Ej. un usuario pasa 10 segundos por día frente a otro usuario que pasa 5 horas por día?
¿Estamos midiendo a los usuarios que pagan y a los que no pagan de manera diferente?

### Determine el tamaño de muestra requerido

A continuación, queremos averiguar cuántos puntos de datos se necesitarán para ejecutar un experimento. Hay cuatro parámetros  que deben configurarse para permitir el cálculo de un tamaño de muestra adecuado:
1. Linea base: una estimación de la métrica que se analiza antes de realizar cambios.

2. Nivel de significación práctica: el cambio mínimo en la linea base que es útil para la empresa, por ejemplo, un aumento en la tasa de conversión del 0,001% puede no valer el esfuerzo requerido para realizar el cambio, mientras que un cambio del 2% será significativo.

3. Nivel de confianza: también llamado nivel de significancia es la probabilidad de que la hipótesis nula sea rechazada cuando no debería ser.

4. Sensibilidad: la probabilidad de que la hipótesis nula no sea rechazada cuando debería ser. 

La linea base se puede estimar utilizando datos históricos, el nivel de importancia práctica dependerá de lo que tenga sentido para la empresa y el nivel de confianza y la sensibilidad se establecen generalmente en 95% y 80% respectivamente, pero se pueden ajustar para adaptarse a diferentes experimentos o necesidades comerciales.

### El diseño del experimento

Armado con métricas bien definidas y requisitos de tamaño de muestra, el paso final en el diseño de un experimento es considerar los elementos prácticos de realizar un experimento. Algunas cosas a tener en cuenta para este paso es decidir:
1. Cuándo debe ejecutarse el experimento (p. Ej., Todos los días o solo días hábiles) y cuánto tiempo debe ejecutarse (p. Ej., 1 día o 1 semana)
2. Cuántos usuarios exponer al experimento por día, donde exponer a menos usuarios significa períodos de tiempo de experimento más largos
3. Cómo contabilizar el efecto de aprendizaje, es decir, permitir que los usuarios aprendan los cambios antes de medir el impacto de un cambio
4. Qué herramientas se utilizarán para capturar datos, como Google Analytics

### Limpiemos los  datos

Al igual que con cualquier análisis de datos tradicional, el primer paso es limpiar los datos recopilados. Este paso es importante para garantizar que los datos se alineen con las definiciones de las métricas determinadas durante el diseño del experimento.

Debe comenzar por ejecutar algunas comprobaciones basicas para asegurarse de que los resultados del experimento reflejen el diseño inicial, p. Ej. aproximadamente un número igual de usuarios vieron la página de destino antigua y la nueva o están las tasas de conversión en el ámbito de las posibilidades.

Este dataset tiene un problema, algunos usuarios de ambos grupos estaban expuestos a la página de destino incorrecta. El grupo de control debería haber estado expuesto a la página old_page, mientras que el grupo de tratamiento debería haber estado expuesto a new_page. 

```python
# algunos miembros del grupo de control vieron new_page y algunos grupos de tratamiento vieron old_page - elimine estas instancias

mask1 = (df["group"] == "control") & (df["landing_page"] == "new_page")
index_to_drop1 = df[mask1].index
df = df.drop(index_to_drop1)

mask2 = (df["group"] == "treatment") & (df["landing_page"] == "old_page")
index_to_drop2 = df[mask2].index
df = df.drop(index_to_drop2)

print(df.shape)
df["group"].value_counts()
```

```python
# Verifiquemos los duplicados. 
print(df["user_id"].count())
print(df["user_id"].nunique())
```

Son pocos asi que los eliminaremos

```python
# Eliminemos usuarios duplicados.
df.drop_duplicates(subset ='user_id',keep ='first',inplace = True)
```

Comprobemos que el tamaño de la muestra sea lo suficientemente grande
Como se menciona en “Diseño de una prueba A / B, , hay cuatro parámetros de prueba para establecer: la linea base, el nivel de significación práctica, el nivel de confianza y la sensibilidad. 

Usemos la probabilidad del grupo de control como una forma de medir el nivel de significancia de la línea de base y establecer el nivel de significancia práctica, el nivel de confianza y la sensibilidad al 1%, 95% y 80% respectivamente. 

Usando estos valores, calculemos el tamaño de muestra mínimo requerido para cada grupo de prueba para asegurarnos de que hay datos suficientes para sacar conclusiones estadísticamente sólidas.

```python
# Identificamos si pertence al grupo de contrl.
mask = (df["group"] == "control")
# Contamos los convertidos
conversions_control = df["converted"][mask].sum()
# Frente al total del grupo de control
total_users_control = df["converted"][mask].count()

# Lo mismo para el caso de los usuarios del grupo treatement
mask = (df["group"] == "treatment")
conversions_treatment = df["converted"][mask].sum()
total_users_treatment = df["converted"][mask].count()

# Esta equilibrada la cantidad de usuarios
print("Separacion de usuarios de control que vieron la página anterior frente a los usuarios de treatment que vieron la página nueva: ", 
          round(total_users_control / df["converted"].count() * 100, 2), "% ",
          round((total_users_treatment / df["converted"].count()) * 100, 2), "%")

#contar el número de usuarios que realizaron conversiones en cada grupo
print("Número de usuarios de control que realizaron conversiones en la página anterior: ", conversions_control)
print("Porcentaje de usuarios de control que realizaron una conversión: ", round((conversions_control / total_users_control) * 100, 2), "%")

mask = (df["group"] == "treatment")
print("Número de usuarios del treatment que se convirtieron en una página nueva: ", conversions_treatment)
print("Porcentaje de usuarios del treatment que se convirtieron: ", round((conversions_treatment/ total_users_treatment) * 100, 2), "%")
```

**Power Analysis**: El poder estadístico, o el poder de una prueba de hipótesis, es la probabilidad de que la prueba rechace correctamente la hipótesis nula.

Es decir, la probabilidad de un resultado positivo verdadero. Solo es útil cuando se rechaza la hipótesis nula.

El poder estadístico es una pieza de un rompecabezas que tiene cuatro partes relacionadas; son:

Tamaño del efecto(effect size): La magnitud cuantificada de un resultado presente en la población. El tamaño del efecto se calcula utilizando una medida estadística específica, como el coeficiente de correlación de Pearson para la relación entre variables o la d de Cohen para la diferencia entre grupos.

Tamaño de la muestra(Sample Size). El número de observaciones en la muestra.
Significancia (Significance). El nivel de significancia o nivel de confianza usado en la prueba estadística. A menudo se establece en 5% o 0.05.

Poder estatico. La probabilidad de aceptar la hipótesis alternativa si es cierta.

Las cuatro variables están relacionadas. Por ejemplo, un tamaño de muestra más grande puede hacer que un efecto sea más fácil de detectar y la potencia estadística se puede aumentar en una prueba aumentando el nivel de significancia. Usualmente se utiliza el 80%


Es un aspecto importante del diseño experimental. Nos permite determinar el tamaño de muestra necesario para detectar un efecto de un tamaño determinado con un grado de confianza determinado. Por el contrario, nos permite determinar la probabilidad de detectar un efecto de un tamaño dado con un nivel de confianza dado, bajo restricciones de tamaño de muestra. Si la probabilidad es inaceptablemente baja, sería prudente modificar o abandonar el experimento.

```python
# Definamos o calculemos los parametros necesarios para definir el tamano de la muestra. 
baseline_rate = conversions_control / total_users_control
practical_significance = 0.01 # Definida por el usuario. 
confidence_level = 0.05 # La define el usuario, usualmente 95%. Significancia
sensitivity = 0.8 # Definida por el usuario, usualmente 0.8. Statistical Power

effect_size = sms.proportion_effectsize(baseline_rate, baseline_rate + practical_significance)
# Vamos a calcular el tamano de la muestra de acuerdo con los parametros definidos. 
sample_size = sms.NormalIndPower().solve_power(effect_size = effect_size, power = sensitivity, 
                                               alpha = confidence_level, ratio=1)
print("Tamano de la poblacion requerida: ", round(sample_size), " por grupo")
```

El tamaño de muestra mínimo de 17.209 es ligeramente menor que los 17.489 usuarios en el grupo de control y 17.264 usuarios en el grupo de tratamiento y, por lo tanto, es suficiente para realizar la prueba de hipótesis.

```python
mask = (df["group"] == "control")
conversions_control = df["converted"][mask].sum()
total_users_control = df["converted"][mask].count()

mask = (df["group"] == "treatment")
conversions_treatment = df["converted"][mask].sum()
total_users_treatment = df["converted"][mask].count()

prob_pooled = (conversions_control + conversions_treatment) / (total_users_control + total_users_treatment)
```

1. Pooled probability: $$ \hat{P}_{pool} = \frac{X_{control} + X_{experiment}}{Users_{control}+ Users_{experiment}}$$

2. Pooled Standard Error: $$ SE_{pool} = \sqrt{\hat{P}_{pool}*(1-\hat{P}_{pool})*(\frac{1}{Users_{control}}+\frac{1}{Users_{experiment}}))}$$

```python
se_pooled = math.sqrt(prob_pooled * (1 - prob_pooled) * (1 / total_users_control + 1 / total_users_treatment))
z_score = st.norm.ppf(1 - confidence_level / 2)
margin_of_error = se_pooled * z_score

# La diferencia estimada entre la probabilidad de conversiones en el experimento y los grupos de control
d_hat = (conversions_treatment / total_users_treatment) - (conversions_control / total_users_control)

# Probar si podemos rechazar la hipótesis nula
lower_bound = d_hat - margin_of_error
upper_bound = d_hat + margin_of_error

if practical_significance < lower_bound:
    print("Se rechaza la hipotesis nula. ")
else: 
    print("No se rechaza la hipotesis nula. ")
    
print("El límite inferior del intervalo de confianza es", round(lower_bound * 100, 2), "%")
print("El límite superior del intervalo de confianza es ", round(upper_bound * 100, 2), "%")
```

En este ejemplo, el intervalo de confianza está entre -0,39% y 0,08%. Dado que el nivel mínimo definido en el nivel de significancia práctica es del 1%, solo podríamos rechazar la hipótesis nula si el límite inferior del intervalo de confianza estuviera por encima del 1%. Por lo tanto, no podemos rechazar la hipótesis nula y concluir qué página de destino genera más conversiones.
