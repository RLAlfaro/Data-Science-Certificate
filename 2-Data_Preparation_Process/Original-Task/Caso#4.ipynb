### Extractor de informacion de Boletas y procesamiento

El proyecto de ahora consiste en crear una base de datos en base a un monton de boletas del SII y verificar si el monto es correcto segun lo requerido, informacion que supuestamente extrae de un txt... en este caso, solo tiene una funcion completada.

```python
import PyPDF2
import pandas as pd
import re
from datetime import date, datetime

import os
```

```python
def preparar_db(df, mainpath, carpeta, carpeta_boletas):
    origen = carpeta
    docdir = os.path.join(mainpath + '/' + carpeta + '/' + carpeta_boletas + '/')
    files = os.listdir(docdir)
    for file in files:
        if file.endswith('.pdf'):
            pdfFileObject = open(mainpath + '/' + carpeta + '/' + carpeta_boletas + '/' + file, 'rb')
            print(pdfFileObject)
            pdfReader = PyPDF2.PdfFileReader(pdfFileObject, strict=False,overwriteWarnings=False)
            # print(" No. Of Pages :", pdfReader.numPages)
            pdfReader.numPages
            pageObject = pdfReader.getPage(0)
            # print(pageObject.extractText())
            text = pageObject.extractText()
            try:
                nombre = re.search('(.+?)BOLETA DE HONORARIOSELECTRONICAN', text).group(1)
            except:
                nombre = "Nameless"
            try:
                rut = re.search('RUT: (.+?)GIRO', text).group(1)
            except:
                rut = "0"
            try:
                numero = re.search('° (.+?)RUT: ', text).group(1)
            except:
                numero = "1"
            try:
                monto = re.search('Total Honorarios (.+?)11.5', text).group(1)[2:]
            except:
                monto = "1"
            try:
                fecha_boleta = re.search('Hora Emisión: (.+?) ', text).group(1)
            except:
                fecha_boleta = "01/01/1900"
            rut_no_dot = rut.replace(".", "")
            monto_no_dot = monto.replace(".", "")
            print(rut)
            pdfFileObject.close()
            df = df.append({'file_name': file,
                            'Carpeta' : origen,
                            'Nombre_Persona': nombre, 
                            'RUT_Persona': rut, 
                            'Numero_Bol': numero, 
                            'Monto_Bol': monto,
                            'Rut_sin_punto': rut_no_dot,
                            'Monto_sin_punto': monto_no_dot,
                            'Fecha_Bol': fecha_boleta}, ignore_index=True)
    # No_dot = []
    return df
```
