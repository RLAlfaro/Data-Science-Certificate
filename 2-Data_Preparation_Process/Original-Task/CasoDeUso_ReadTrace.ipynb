### Leer un archivo de un sensor (geofono, https://es.wikipedia.org/wiki/Ge%C3%B3fono) y convertir en un dataframe de panda. Se devuelven las tres componentes de velocidades que mide el sensor. Oportunidades de mejora: agregar la columna tiempo para realizar otras aplicaciones (como por ejemplo, sincronizar sensores).

```python
def ReadTrace(filename, length_header) :
  import numpy as np
  import pandas as pd

  # leer archivo
  content_file = np.loadtxt(filename, skiprows = length_header, encoding='windows-1252')

  # definir nombre de las columnas
  c = ['NS [mm/s]',       'EW [mm/s]',        'Z [mm/s]',        'Time_mar [count]']

  # convertir a Pandas Dataframe
  data_rec = pd.DataFrame.from_records(content_file, columns = c) 

  #definir tiempo de inicio y fin
  with open(filename, 'r', encoding='windows-1252') as td:
    for line in td:
      # print(line)
      # encontrar lineas con metadata de interés
      if 'Start recording:' in line :
        start_rec = line
      elif 'Sampling rate:' in line :
        smp_rate = line

  #crear una columna con tiempo 

  #extraer 
  return data_rec
```
